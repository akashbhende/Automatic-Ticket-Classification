Preprocessing the text

```python
from gensim.parsing.preprocessing import remove_stopwords
```

```python
words = set(nltk.corpus.words.words())
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
words_to_remove = ["hi", "hello", "thank", "thanks", "regards", "welcome","good afternoon","good morning","thank you","good morning","good evening"] # List of words to be removed from the data Greetings
```

```python
def preprocess_text(txt):
  txt = txt.lower()    #Converting the text into lower case
  txt = re.sub(r'\S+@\S+', ' ', txt)  #Removing email address
  txt = re.sub('[^a-z ]+', ' ', txt)  #Removing specaial characters and numbers
  txt = txt.replace('x d'," ")  #It is observable that many text contains the charcters 'x d' repeatedly, so removing it
  txt = txt.replace('xd'," ")
  txt = txt.replace('zz'," ") #It is also seen that repeated use of 'zz' , so removing it.
  txt = ' '.join(txt.split()) #Removing extra spaces
  txt = remove_stopwords(txt)  #Calling a function to remove stopswords
  txt = " ".join(gw for gw in nltk.wordpunct_tokenize(txt) if gw.lower() not in words_to_remove)   #Removing greetings
  txt = ps.stem(txt) #Stemming
  return txt
```

```python
df_ticket["Ticket description"] = df_ticket["Ticket description"].apply(preprocess_text)
```

```python
def preprocess_text(txt):
  txt = txt.lower()    #Converting the text into lower case
  txt = re.sub(r'\S+@\S+', ' ', txt)  #Removing email address
  txt = re.sub('[^a-z ]+', ' ', txt)  #Removing specaial characters and numbers
  txt = txt.replace('x d'," ")  #It is observable that many text contains the charcters 'x d' repeatedly, so removing it
  txt = txt.replace('xd'," ")
  txt = txt.replace('zz'," ") #It is also seen that repeated use of 'zz' , so removing it.
  txt = ' '.join(txt.split()) #Removing extra spaces
  txt = remove_stopwords(txt)  #Calling a function to remove stopswords
  txt = " ".join(gw for gw in nltk.wordpunct_tokenize(txt) if gw.lower() not in words_to_remove)   #Removing greetings
  txt = " ".join(w for w in nltk.wordpunct_tokenize(txt) if w.lower() in words)
  txt = ps.stem(txt) #Stemming
  return txt
```
